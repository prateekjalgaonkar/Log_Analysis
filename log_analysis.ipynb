### Analyzing the log and investigating if there are any patterns

```python
(df['Date'].value_counts(normalize=True)*100).sort_index()
```

__We can observe there are only two dates data here, i.e 23Dec2017 and 24Dec2017 of which 89% of the data is dated to 23rd Dec__

```python
(df['Hours'].value_counts(normalize=True)*100).sort_index()
```

__Observation__
- We can observe at 22 and 23 hours the maximum process are running, may be because it is end of the day. So this data is dependant on some of those attributes where the processes runs in EOD

```python
(df['Process_name'].value_counts(normalize=True)*100).sort_index()
```

__Here we observed that value count for `Step_LSC`, `Step_SPUtils` and `Step_ExtSDM` is maximum , covering around `85%` of the data. We will try to find pattern over the above three attributes.__

```python
(df['Num_id'].value_counts(normalize=True)*100).sort_index()
```

__Since we have only one type of data in column `Num_id`, we will drop this__

```python
df= df.drop('Num_id',axis=1)
```

```python
df.loc[(df['Process_name'] == 'Step_LSC')].head(10)
```

```python
df.loc[(df['Process_name'] == 'Step_ExtSDM')].head(10)
```

__Observation__
- We can find a pattern here and observe that calories calculation and altitude calculation process is repeated back to back after a set interval.
- There is an interval of 6-8 processes before the calories and altitude calculation takes place.


```python
df.loc[(df['Process_name'] == 'Step_SPUtils')].head(20)
```

__Observation__
- We can find a pattern here and observe that gettodaydetailssteps and settodaydetailssteps is repeated back to back after a set interval.
- Here also there is an interval of 6-8 processes before the set and get processes takes place.
